## Model Training and Evaluation

This section trains two different regression models and evaluates their performance.

### Splitting Data into Features and Target

Separates the dataset into features (X) and the target variable (y), which is 'Aggregate rating'.

```python
# Separate features (X) and target variable (y)
x = restaurant_df.drop('Aggregate rating', axis=1)
y = restaurant_df['Aggregate rating']
```

### Splitting Data into Training and Testing Sets

Splits the features and target variable into training and testing sets.

```python
# Split data into training and testing sets (80% train, 20% test)
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

### Verifying Shape of Training and Testing Sets

Displays the shapes of the training and testing sets.

```python
# Verify the shapes of the training and testing sets
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

### Training Linear Regression Model

Trains a Linear Regression model on the training data.

```python
# Train a Linear Regression model
from sklearn.linear_model import LinearRegression
linreg = LinearRegression()
linreg.fit(X_train,y_train)
linreg_pred=linreg.predict(X_test)
```

### Evaluating Linear Regression Model

Evaluates the performance of the Linear Regression model using MAE, MSE, and R2 score.

```python
# Evaluate Linear Regression model performance using MAE, MSE, and R2 score
from sklearn.metrics import mean_squared_error, r2_score, accuracy_score, mean_absolute_error

linreg_mae = mean_absolute_error(y_test, linreg_pred)
linreg_mse = mean_squared_error(y_test, linreg_pred)
linreg_r2 = r2_score(y_test, linreg_pred)

print(f"MAE of the linear regression model is: {linreg_mae:.2f}")
print(f"MSE of the linear regression model is: {linreg_mse:.2f}")
print(f"R2 score of the linear regression model is: {linreg_r2:.2f}")
```

### Training Decision Tree Regression Model

Trains a Decision Tree Regression model on the training data.

```python
# Train a Decision Tree Regression model
from sklearn.tree import DecisionTreeRegressor
dec_tree = DecisionTreeRegressor()
dec_tree.fit(X_train, y_train)
dec_tree_pred = dec_tree.predict(X_test)
```

### Evaluating Decision Tree Regression Model

Evaluates the performance of the Decision Tree Regression model using MAE, MSE, and R2 score.

```python
# Evaluate Decision Tree Regression model performance using MAE, MSE, and R2 score
from sklearn.metrics import mean_squared_error, r2_score, accuracy_score, mean_absolute_error

dtree_mae = mean_absolute_error(y_test, dec_tree_pred)
dtree_mse = mean_squared_error(y_test, dec_tree_pred)
dtree_r2 = r2_score(y_test, dec_tree_pred)

print(f"MAE of the decision tree model is: {dtree_mae:.2f}")
print(f"MSE of the decision tree model is: {dtree_mse:.2f}")
print(f"R2 score of the decision tree model is: {dtree_r2:.2f}")
```
