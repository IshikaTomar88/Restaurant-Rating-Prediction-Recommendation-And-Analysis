# Geographical Analysis

This section focuses on the geographical aspects of the restaurant data, including visualizing locations and analyzing distributions by city.

### Visualize restaurant locations
This code visualizes the restaurant locations on an interactive map using the `folium` library. Markers are added for each restaurant, and a `MarkerCluster` is used to group nearby markers for better visualization.

```python
from google.colab import files

files.download('restaurants_map.html')
```

```python
# Visualize restaurant locations on a map using Folium
import folium
from folium.plugins import MarkerCluster

# Calculate the mean latitude and longitude for centering the map
latitudes = restaurant_df['Latitude'].mean()
longitudes = restaurant_df['Longitude'].mean()

# Create a Folium map centered at the mean coordinates
map = folium.Map(location=[latitudes, longitudes], zoom_start=12)

# Create a MarkerCluster to group nearby markers
marker_cluster = MarkerCluster().add_to(map)

# Iterate through each row in the dataframe to add markers for each restaurant
for index, row in restaurant_df.iterrows():
    # Create popup text with restaurant information
    popup_text = f"<b>{row['Restaurant Name']}</b><br>" \
                 f"City: {row['City']}<br>" \
                 f"Rating: {row['Aggregate rating']}<br>" \
                 f"Cuisines: {row['Cuisines']}"
    # Add a marker to the map
    folium.Marker(
        location=[row['Latitude'], row['Longitude']],
        popup = popup_text,
        icon=folium.Icon(color='red', icon='info-sign')
        ).add_to(marker_cluster)

# Add a layer control to the map
folium.LayerControl().add_to(map)

# Save the map as an HTML file
map.save('restaurants_map.html')
# Display the map
display(map)
```

# Restaurant Concentration by City

This section analyzes the concentration of restaurants in different cities.

### Count of Restaurant in City
This code groups the DataFrame by 'City' and counts the number of restaurants in each city using `.size()`. The results are sorted to identify cities with the highest restaurant counts.

```python
# Group the DataFrame by 'City' and count the number of restaurants in each city
grouped_by_city = restaurant_df.groupby('City').size().reset_index(name='Count')

# Sort the cities by the restaurant count in descending order and take the top 15
grouped_by_city = grouped_by_city.sort_values("Count", ascending=False).head(15)
# Display the result
grouped_by_city
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set the figure size
plt.figure(figsize=(10, 6))
# Create a bar plot of restaurant count by city
sns.barplot(x='City', y='Count', data=grouped_by_city, palette="flare")
plt.xlabel('City')
plt.ylabel('Count')
plt.title('Count of Restaurants by City')
# Rotate x-axis labels for better readability
plt.xticks(rotation=90)
# Show the plot
plt.show()
```

## Restaurant Ratings Analysis by City

This section analyzes the average ratings of restaurants in different cities.

### Average Ratings by City
This code calculates the average rating for each city by grouping the DataFrame by 'City' and applying the `.mean()` aggregation to the 'Aggregate rating' column.

```python
# Calculate the average rating for each city
average_rating_by_city = restaurant_df.groupby('City')['Aggregate rating'].mean().reset_index()

# Print the average rating by city
print("Average Rating by City:")
print(average_rating_by_city)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate the average rating for each city and rename the column
average_rating_by_city = restaurant_df.groupby('City')['Aggregate rating'].mean().reset_index(name = "Average Rating")
# Sort by average rating in descending order
average_rating_by_city = average_rating_by_city.sort_values("Average Rating", ascending=False)

# Select the top 10 cities
top_cities = average_rating_by_city.head(10)

# Set the figure size
plt.figure(figsize=(10, 6))
# Create a bar plot of the top 10 cities by average rating
sns.barplot(x='City', y='Average Rating', data=top_cities, palette="rocket")
plt.xlabel('City')
plt.ylabel('Average Rating')
plt.title('Top 10 Cities by Average Rating')
# Rotate x-axis labels for better readability
plt.xticks(rotation=90)
# Show the plot
plt.show()
```

## Price Range Analysis by City

This section analyzes the most common price range of restaurants in different cities.

### Price Range Analysis by City
This code finds the most common price range for each city by grouping the DataFrame by 'City' and using the `.mode()` aggregation on the 'Price range' column.

```python
# Find the most common price range by city'
common_price_range_by_city = restaurant_df.groupby('City')['Price range'].agg(lambda x: x.mode()[0] if not x.mode().empty else 'No Data')
# Print a descriptive header
print("Most Common Price Range by City:")
# Display the resulting Series
print(common_price_range_by_city)
```

```python
# Set the figure size for better readability
plt.figure(figsize=(10, 6))

# Create a bar plot for common price range by city
for city, price_range in common_price_range_by_city.items():
    # Create a bar for each city with its common price range
    plt.bar(city, price_range, label=price_range)

# Label for the x-axis
plt.xlabel('City')
# Label for the y-axis
plt.ylabel('Common Price Range')
# Title of the plot
plt.title('Common Price Range by City')
# Rotate x-axis labels for readability
plt.xticks(rotation=90)
# Display the plot
plt.show()
```

## Cuisine Diversity by City

This section analyzes the diversity of cuisines offered in different cities.

### Popular Cuisines in City
This code calculates the number of unique cuisines offered in each city by grouping the DataFrame by 'City' and applying a lambda function to count the unique values in the 'Cuisines' column using `len(set(x))`.

```python
# Group the DataFrame by 'City'
grouped_by_city = restaurant_df.groupby('City')

# Calculate the number of unique cuisines for each city
cuisine_diversity = grouped_by_city['Cuisines'].apply(lambda x: len(set(x)))

# Print the cuisine diversity for each city
print(cuisine_diversity)
```
